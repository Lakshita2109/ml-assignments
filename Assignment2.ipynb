```python
relevant = customers[['CustomerID', 'City', 'Education', 'Gender', 'MaritalStatus','NumberCarsOwned', 'NumberChildrenAtHome','YearlyIncome']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
relevant["Income_norm"] = scaler.fit_transform(relevant[["YearlyIncome"]])
```
